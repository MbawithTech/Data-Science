```python
import numpy as np
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("sales_data.csv")
```

```python
df.head()
```

```python
df.drop(columns=["Units_sold"], axis=1, inplace=True)
```

```python
df.values
```

```python
df[(df["1_Star_Rating"] == 0) & (df["2_Star_Rating"] == 0)]
```

```python
ratings_ratio = []

for row in df.values:
  if (row[4] + row[5] == 0):
    if (row[7] + row[8] == 0):
      ratings_ratio.append(0.0)
    else:
      ratings_ratio.append(-9999)
  
  else:
    ratings_ratio.append((int(row[7]) + int(row[8])) / (int(row[4]) + int(row[5])))
```

```python
ratings_ratio
```

```python
max_ratio = max(ratings_ratio)
max_ratio
```

```python
for index in range(len(ratings_ratio)):
  if ratings_ratio[index] == -9999:
    ratings_ratio[index] = max_ratio
  else:
    pass
```

```python
ratings_ratio
```

```python
df["Good_by_bad_rating"] = ratings_ratio
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(labels=["1_Star_Rating", "2_Star_Rating","3_Star_Rating","4_Star_Rating", "5_Star_Rating"], axis=1, inplace = True)
```

```python
df["Segment"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df[["Segment"]] = df[["Segment"]].apply(LabelEncoder().fit_transform)
```

```python
df.head()
```

Perform Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(labels=["Units_sold>1000"], axis=1)
```

```python
y = df["Units_sold>1000"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT_model = DecisionTreeClassifier(random_state=42)
```

```python
DT_model.fit(X_train, y_train)
```

```python
y_train_pred = DT_model.predict(X_train)
y_pred = DT_model.predict(X_test)
```

Evaluating the model

```python
from sklearn.metrics import classification_report
```

```python
class_report_test = classification_report(y_train, y_train_pred)
print("The training report is:")
print(class_report_test)
```

```python
test_set_report = classification_report(y_test, y_pred)
print("Test report is:")
print(test_set_report)
```

Calculating Feature Importance

```python
features_imp = DT_model.feature_importances_
```

```python
imp = pd.DataFrame(
  {"Features": X_train.columns,
   "Importance": np.round(features_imp,2)
   }
)
```

```python
imp = imp.sort_values(by="Importance", ascending=False)
```

```python
imp
```

```python
imp["cumulative_sum"] = imp.Importance.cumsum()
```

```python
imp.head()
```

```python
drop_records = imp[imp["cumulative_sum"] > 0.90]["Features"].to_list()
```

```python
drop_records
```

```python
X_train_new = X_train.drop(labels=drop_records, axis=1)
```

```python
X_test_new = X_test.drop(labels=drop_records, axis=1)
```

Making predictions using new model

```python
DT_model.fit(X_train_new, y_train)
```

```python
y_train_new_pred = DT_model.predict(X_train_new)
y_pred_new = DT_model.predict(X_test_new)
```

```python
print(classification_report(y_train, y_train_new_pred))
```

```python
print(classification_report(y_test, y_pred_new))
```

Hyperparamter tuning

```python
#Storing the depth of current model
depth = DT_model.get_depth()
depth
```

```python
#creating a list of depth values
max_depth_list  = list(range(depth,0,-3))
max_depth_list
```

```python
from sklearn.metrics import f1_score
```

```python
train_f1_scores = []
test_f1_scores = []

# collecting training and test f1 scores
for depth in max_depth_list:
  D_model = DecisionTreeClassifier(max_depth=depth, random_state=42)

  #fitting the model on train data
  D_model.fit(X_train_new, y_train)

  #predicting the model
  y_train_pred_f1 = D_model.predict(X_train_new)
  y_pred_f1  = D_model.predict(X_test_new)

  #calculating f1_score
  train_f1_scores.append(f1_score(y_train, y_train_pred_f1))
  test_f1_scores.append(f1_score(y_test,y_pred_f1))
```

```python
train_f1_scores
```

```python
print(len(max_depth_list))
print(len(train_f1_scores))
print(len(test_f1_scores))
```

```python
depth_df = pd.DataFrame(
  {
    "max_depth_value" : max_depth_list,
    "train_f1_scores" : np.round(train_f1_scores,3),
    "test_f1_scores" : np.round(test_f1_scores,3)
  }
)
```

```python
depth_df
```

```python
depth_df["f1_score_gap"] = depth_df["train_f1_scores"] - depth_df["test_f1_scores"]
```

```python
depth_df
```

From the above table we can see that for max_depth = 6 we get the best performing model for both train and test data
