Calculating VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calculate_vif(data):
  vif_df = pd.DataFrame()
  vif_df["Variable"] = data.columns
  vif_df["VIF"] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]
  vif_df["VIF"] = pd.to_numeric(vif_df["VIF"])
  return vif_df

result_vif = calculate_vif(X1)
```

```python
result_vif["VIF"] = pd.to_numeric(result_vif["VIF"], errors='coerce')
```

```python
result_vif["VIF"] = result_vif["VIF"].apply(lambda x: np.round(x,3))
```

```python
result_vif
```

```python
df.columns
```

```python
from math import sqrt
```

```python
df.info()
```

```python
# Converting latitude and longitude values to radians
df["pickup_latitude_radian"] = df["pickup_latitude"].apply(lambda x: np.radians(x))
df["pickup_longitude_radian"] = df["pickup_longitude"].apply(lambda x: np.radians(x))
df["dropoff_latitude_radian"] = df["dropoff_latitude"].apply(lambda x: np.radians(x))
df["dropoff_longitude_radian"] = df["dropoff_longitude"].apply(lambda x: np.radians(x))
```

```python
df.head()
```

```python
df["delta_latitude"] = df["dropoff_latitude_radian"] - df["pickup_latitude_radian"]
df["delta_longitude"] = df["dropoff_longitude_radian"] - df["pickup_longitude_radian"]
```

```python
df.head()
```

```python
df["haversine_a"] = ((np.sin(df["delta_longitude"] / 2))**2)*(np.cos(df["dropoff_latitude_radian"]))*(np.cos(df["pickup_latitude_radian"])) 
+ (np.sin(df["delta_latitude"] / 2))**2
```

```python
df["haversine_a"].head()
```

```python
df["haversine_c"] = 2 * np.arcsin(np.sqrt(df["haversine_a"]))
```

```python
df["haversine_c"].head()
```

```python
df["haversine_distance_km"] = 6371*df["haversine_c"]
```

```python
df["haversine_distance_km"].head()
```
