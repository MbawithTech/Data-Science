```python
import numpy as np
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("sales_data.csv")
```

```python
df.head()
```

```python
#Good_by_bad_ratio
ratings_ratio = []

for row in df.values:
  if (row[4] + row[5] == 0):
    if (row[7] + row[8] == 0):
      ratings_ratio.append(0.0)
    else:
      ratings_ratio.append(-9999)
  
  else:
    ratings_ratio.append((int(row[7])+int(row[8])) / (int(row[4]) + int(row[5])))
```

```python
ratings_ratio
```

```python
for index in range(len(ratings_ratio)):
  if ratings_ratio[index] == -9999:
    ratings_ratio[index] = max(ratings_ratio)
    ratings_ratio[index] = round(ratings_ratio[index], 3)
  else:
    ratings_ratio[index] = round(ratings_ratio[index], 3)
```

```python
ratings_ratio
```

```python
df["Good_by_bad_ratings"] = ratings_ratio
```

```python
df.head()
```

```python
df_new = df.drop(labels=["1_Star_Rating", "2_Star_Rating", "3_Star_Rating", "4_Star_Rating", "5_Star_Rating", "Units_sold>1000"], axis=1)
```

```python
df_new.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data_encoder = LabelEncoder()
```

```python
df_new[["Segment"]] = df_new[["Segment"]].apply(data_encoder.fit_transform)
```

```python
df_new.head()
```

Building the model

```python
X = df_new.drop(labels=["Units_sold"], axis=1)
y = df_new["Units_sold"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
DTR_model = DecisionTreeRegressor(random_state=42)
```

```python
DTR_model.fit(X_train, y_train)
```

```python
y_train_pred = DTR_model.predict(X_train)
y_pred = DTR_model.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
print(f'Training dataset_RMS: {root_mean_squared_error(y_train, y_train_pred)}')
print(f'Test data RMS: {root_mean_squared_error(y_test, y_pred)}')
```

```python
train_data_df = pd.DataFrame(
  {
    "Original_train_Y" : y_train.tolist(),
    "Predicted_train_Y" : y_train_pred.tolist()
  }
)
```

```python
train_data_df.head()
```

```python
from sklearn.metrics import r2_score
```

```python
print(f'Training dataset R2_score: {r2_score(y_train, y_train_pred)}')
print(f'Test dataset R2_score: {r2_score(y_test, y_pred)}')
```
