```python
import numpy as np
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("sales_data.csv")
```

```python
#Good_by_bad_ratio
ratings_ratio = []

for row in df.values:
  if (row[4] + row[5] == 0):
    if (row[7] + row[8] == 0):
      ratings_ratio.append(0.0)
    else:
      ratings_ratio.append(-9999)
  
  else:
    ratings_ratio.append((int(row[7])+int(row[8])) / (int(row[4]) + int(row[5])))
```

```python
for index in range(len(ratings_ratio)):
  if ratings_ratio[index] == -9999:
    ratings_ratio[index] = max(ratings_ratio)
    ratings_ratio[index] = round(ratings_ratio[index], 3)
  else:
    ratings_ratio[index] = round(ratings_ratio[index], 3)
```

```python
df["Good_by_bad_ratings"] = ratings_ratio
```

```python
df_new = df.drop(labels=["1_Star_Rating", "2_Star_Rating", "3_Star_Rating", "4_Star_Rating", "5_Star_Rating", "Units_sold>1000"], axis=1)
```

```python
df.head()
```

```python
df1 = df.drop(labels=["1_Star_Rating", "2_Star_Rating", "3_Star_Rating", "4_Star_Rating", "5_Star_Rating", "Units_sold"], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df1[["Segment"]] = df1[["Segment"]].apply(LabelEncoder().fit_transform)
```

```python
df1.head()
```

```python
X = df1.drop(columns=["Units_sold>1000"], axis=1)
```

```python
y = df1["Units_sold>1000"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=6)
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
#Stratified Splitting
X_train_st, X_test_st, y_train_st, y_test_st = train_test_split(X, y, test_size=0.3, stratify=y, random_state=6)
```

```python
print(y_train_st.value_counts(normalize=True))
print(y_test_st.value_counts(normalize=True))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT_model = DecisionTreeClassifier(max_depth=11, min_samples_leaf=6, random_state=42, class_weight="balanced")
```

```python
DT_model.fit(X_train_st, y_train_st)
```

```python
from sklearn.metrics import f1_score, classification_report
```

```python
y_train_pred_st = DT_model.predict(X_train_st)
y_pred_st = DT_model.predict(X_test_st)
```

```python
f1_score(y_train_st, y_train_pred_st)
```

```python
f1_score(y_test_st, y_pred_st)
```

```python
print(classification_report(y_train_st, y_train_pred_st))
```

```python
print(classification_report(y_test_st, y_pred_st))
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
y_train_pred_rus = DT_model_new.predict(X_train_rus)
y_pred_rus = DT_model_new.predict(X_test)
```

```python
print(classification_report(y_train_rus, y_train_pred_rus))
```

```python
print(classification_report(y_test, y_pred_rus))
```

Oversampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
X_train_ros, y_train_ros = rs_sampler.fit_resample(X_train, y_train)
```
